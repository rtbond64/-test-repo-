```python
#import libraries of earth engine end geemap for python envi
import ee
import geemap
```

```python
# Authentication and initializae ee 
ee.Authenticate()
ee.Initialize(project='ee-st122150')
```

```python
# This is how we write our first Earth Engine Python API Code and get started with collab and Google Earth Engine.
# Import a staelite data in this case, LandSat Satelite image and generate a true and false color composite.
import geemap
Map = geemap.Map()
Map

# Load countries data
countries = ee.FeatureCollection("USDOS/LSIB_SIMPLE/2017"); #This will import the country database
roi = countries.filter(ee.Filter.eq("country_na", "Swaziland")); #This subset into the specific country database

 #import the Landsatelite image collection  Update data, suebet by ROI
landsat = ee.ImageCollection("LANDSAT/LC08/C01/T1") \
  .filterDate('2016-01-01', '2017-01-01') \
  .filterBounds(roi)

#function that run cloud free image
composite = ee.Algorithms.Landsat.simpleComposite(**{
    'collection': landsat,
    'asFloat': True
})
#These are two images: The first one is colorful composite.  rgb standard  Real color
rgbVis = {'bands':["B4", "B3", "B2"], 'min':0, 'max':0.3}
# The second one is false color composite that uses the near infrared band as well.
nirVis = {'bands':["B5", "B4", "B3"], 'min':0, 'max':[0.5, 0.3, 0.3]}

#call composite image  map 
# We use the map.addlayer fuction on earth engine api and clip it by my steady region (roi)
# And provide color parameter (rgbVIS) which is the true color visualization. The second one (nirVis) is near-infrared visualization which is a false color composite.
Map.addLayer(composite.clip(roi), rgbVis, "RGB")
Map.addLayer(composite.clip(roi), nirVis, "False Color")
# Create a map sender object to adjust the zoom level.  that I am interested.
Map.centerObject(roi, 8)
Map
```
